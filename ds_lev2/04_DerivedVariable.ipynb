```python
# pandas - get_dummies()
bike_dum = pd.get_dummies(data=bike, columns=['season'],dtype=int) # one-hot-encoding
bike_dum
```

```python
bike_dum = pd.get_dummies(data=bike, columns=['season'],
                          drop_first=True, dtype=int) # 가변수 (one-hot-encoding에서 변수 하나 제거)
bike_dum
```

```python
abs(bike['temp']-bike['atemp']).mean()
```

```python
bike['date'] = pd.to_datetime(bike['datetime']).dt.date
bike['over25']=np.where(bike['casual']>25,1,0)
bike.groupby(['over25']).head()
```

```python
pd.crosstab(bike['over25'], bike['date']) # 빈도표
```

```python
max_casual_bike = bike.groupby('date')['casual'].max().reset_index()
max_casual_bike = bike.groupby('date')['casual'].max()
max_casual_bike
```

```python
len(max_casual_bike.where(max_casual_bike['casual']>25).dropna())
```

```python
len(max_casual_bike[max_casual_bike>25])
```

```python
bike['hour'] = pd.to_datetime(bike['datetime']).dt.hour
bike
```

```python
reg_mean_bike = bike.groupby('hour')['registered'].mean().reset_index()
reg_mean_bike
```

```python
reg_mean_bike.loc[reg_mean_bike['registered'].idxmax(),]
```
