## 정렬

#### pandas - crosstab()

기본적으로 데이터프레임의 두 변수의 원소 조합 빈도를 확인하는 함수

#### pandas - sort_values()

데이터프레임의 특정 변수를 기준으로 정렬하는 메소드

#### pandas - melt()

wide form 데이터프레임을 long form으로 자료구조를 변환하는 메소드



id_vars 인자에 기준이 되는 변수를 지정하여 처리

#### pandas - pivot()

long form 데이터프레임을 wide form으로 자료구조를 변환하는 메소드



index/columns/values 인자에 각각 대상 변수를 지정하여 출력 데이터 구조 결정

```python
import pandas as pd

dm = pd.read_csv('ex/diamonds.csv')
dm.head(2)
```

```python
pd.crosstab(dm['cut'], dm['clarity'])
```

```python
pd.crosstab(dm['cut'], dm['clarity'], normalize=True)
```

```python
pd.crosstab(dm['cut'], dm['clarity'], normalize=True).round(3)
```

```python
pd.crosstab(dm['cut'], dm['clarity'], normalize=1).round(3)
```

```python
pd.crosstab(dm['cut'], dm['clarity'], normalize=0).round(3)
```

```python
pd.crosstab(dm['cut'], dm['clarity'], values=dm['price'], aggfunc=pd.Series.mean).round(2)
```

```python
dm.groupby(['cut', 'clarity'])['price'].mean().reset_index().head(1)
```

```python
dm_agg = dm.groupby(['cut', 'clarity'])['price'].mean().round(1).reset_index()
dm_agg.head(1)
```

```python
dm_agg.sort_values('price', ascending=False).head(1)
```

```python
dm_agg.sort_values(['cut','clarity'], ascending=[True, False]).head(1)
```

```python
# melt
elec = pd.read_csv('ex/elec_load_2017_7d.csv')
elec.head(2)
```

```python
elec_melt = elec.melt(id_vars=['YEAR', 'MONTH', 'DAY'])
elec_melt.head(2)
```

```python
print(len(elec))
print(len(elec_melt))
```

```python
# pivot
elec_melt.pivot(index=['YEAR', 'MONTH', 'DAY'],
                columns='variable', values='value').reset_index()
```

```python
bike = pd.read_csv('ex/bike.csv')
bike
```

```python
pd.crosstab(bike['holiday'], bike['workingday'], normalize=True)
```

```python
crossed_dm = pd.crosstab(dm['cut'], dm['color']).reset_index()
crossed_dm
```

```python
crossed_dm.melt(id_vars=['cut']).sort_values('value', ascending=False).head(1)
```

```python
dm_price = dm.groupby(['cut','color'])['price'].mean().reset_index()
dm_carat = dm.groupby(['cut','color'])['carat'].mean().reset_index()
print(dm_price.head(3))
print(dm_carat.head(3))
```

```python
price_carat_merged = pd.merge(left=dm_price, right=dm_carat, left_on=['cut','color'], right_on=['cut','color'])
price_carat_merged.head(3)
```

```python
price_carat_merged['price_per_carat'] = price_carat_merged['price']/price_carat_merged['carat']
```

```python
price_carat_merged.sort_values('price_per_carat', ascending=False).head(1)
```

```python
dm_agg = dm.groupby(['cut', 'color'])[['price', 'carat']].mean().reset_index()
dm_agg.head(2)
```

```python
dm_agg['ratio'] = dm_agg['price'] / dm_agg['carat']
dm_agg = dm_agg.sort_values('ratio', ascending=False)
dm_agg.head()
```
