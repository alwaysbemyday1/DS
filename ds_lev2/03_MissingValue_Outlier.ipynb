## 데이터 전처리
### 1️⃣결측치, 이상치
#### 1. 이상치
##### 1) 정의
중심 경향성에서 멀리 떨어진 값
##### 2) 처리 방법
- 절대적 기준 X
- 대표적으로 Carling, Tukey
- 분포 기반으로도 처리 가능(ex. 정규분포 기반 상하위 1% 제거)
#### 2. 결측치
##### 1) 정의
값이 기록되지 않고 비어 있음 (반대는 관측치)
##### 2) 처리 방법
- 절대적 기준 X
- 단순 제거 or 특정값으로 대체
- 결측치 비중이 낮을 경우 단순 제거

```python
df_na = df.head(7)
df_na.notna() # NA가 아닌 것은 True
df_na.isna() # NA인 것(결측치)은 True
df_na.isna().sum() # 각 변수(column)별 결측치 개수
df_na.isna().sum(axis=1) # 각 row별 결측치 개수
```

```python
df_na.fillna(999) # 일괄적으로 결측치 채우기
df_na.fillna(value={'Sepal_Length':999, 'Sepal_Width':888}) # 지정값으로 결측치 채우기
```

```python
df_na['Sepal_Width'].mean() # 결측치를 제외한 Sepal_Width의 평균값
df_na.fillna(value = {'Sepal_Width' : df_na['Sepal_Width'].mean()}) # Sepal_Width 평균값으로 결측치 채우기
```

```python
df_na2 = df.head(7)
df_na2.dropna() # 결측치 제거
df_na2.dropna(how='all') # row의 모든 값이 결측치면 제거
df_na2.iloc[:,:-1].dropna(how='all') # 마지막 column 제외한 row의 모든 값이 결측치면 제거
```

```python
print(df['Sepal_Length'].max())
print(df['Sepal_Length'].min())
print(df['Sepal_Length'].mean())
print(df['Sepal_Length'].median()) # 중위수
print(df['Sepal_Length'].quantile()) # 사분위수 (0.5=중위수)
print(df['Sepal_Length'].quantile(q = 0.25)) # 제 1 사분위수
print(df['Sepal_Length'].quantile(q = 0.75)) # 제 3 사분위수
```

```python
df.iloc[:,:-1].isna().sum().sum() # 수치형 변수의 결측치 총 개수
filled_df = df.fillna(value={'Sepal_Width':df['Sepal_Width'].mean()})
((filled_df['Sepal_Width']-filled_df['Sepal_Width'].mean())**2).sum()/filled_df.shape[0] # 분산 힘들게 구하는 법
filled_df['Sepal_Width'].var() # 분산 쉽게 구하는 법
```

```python
iris = pd.read_csv('example/iris.csv')
i_mean = iris['Sepal.Length'].mean()
i_std = iris['Sepal.Length'].std() # 표준편차

c1 = iris['Sepal.Length'] < i_mean - 1.5 * i_std # 표준편차 1.5 미만
c2 =  iris['Sepal.Length'] > i_mean + 1.5 * i_std # 표준편차 1.5 초과
df_out = df.loc[c1|c2,] # 표준편차 1.5 이상의 이상치 제거
print(len(df_out))
df_out
```
